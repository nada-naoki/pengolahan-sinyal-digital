```python
# Mendefinisikan sinyal diskrit x[n]
x = [1, 2, 3, 4, 5]

# Menghitung energi sinyal
energi = 0
for nilai in x:
    energi += nilai ** 2

# Menampilkan hasil energi
print("Energi dari sinyal diskrit x[n] adalah:", energi)
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 1, 1000)
x = 5 * np.sin(4 * np.pi * t)

plt.plot(t, x)
plt.title("Sinyal Sinusoidal")
plt.xlabel("Waktu (detik)")
plt.ylabel("Amplitudo")
plt.grid(True)
plt.show()
```

```python
def convolution(x, h):
    x_len = len(x)
    h_len = len(h)
    y_len = x_len + h_len - 1
    y = np.zeros(y_len)
    
    for n in range(y_len):
        for k in range(h_len):
            if (n - k) >= 0 and (n - k) < x_len:
                y[n] += x[n - k] * h[k]
    return y

x = np.array([1, 2, 3])
h = np.array([0, 1, 0.5])
y = convolution(x, h)
print(f"Hasil konvolusi: {y}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Inisialisasi parameter
n = range(0, 11)  # n dari 0 hingga 10
x = []  # List untuk menyimpan nilai x[n]

# Menghitung x[n] menggunakan loop for
for i in n:
    x_n = 3 * np.exp(0.5 * i)  # x[n] = 3 * exp(0.5 * n)
    x.append(x_n)

# Memplot sinyal x[n]
plt.stem(n, x, use_line_collection=True)
plt.title(r'Diskrit Sinyal Eksponensial $x[n] = 3e^{0.5n}$')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.grid(True)
plt.show()
```

```python
def LTI_system(x, y_initial):
    y = np.zeros(len(x))
    y[0] = 0.5 * y_initial + x[0]
    for n in range(1, len(x)):
        y[n] = 0.5 * y[n-1] + x[n]
    return y

x = np.array([1, 2, 3, 4, 5])
y_initial = 0
y = LTI_system(x, y_initial)
print(f"Output dari sistem: {y}")
```
