```python
import numpy as np
import matplotlib.pyplot as plt

# Paramater sinyal
A = 1      # Amplitudo
f = 5      # Frekuensi (Hz)
T = 1/100  # Periode sampling untuk sinyal diskrit (100 Hz)
t = np.linspace(0, 1, 1000)  # Waktu untuk sinyal kontinu (1 detik)

# Sinyal waktu kontinu
x_continuous = A * np.sin(2 * np.pi * f * t)

# Sinyal waktu diskrit (sampel setiap 10 ms)
n = np.arange(0, 1, T)
x_discrete = A * np.sin(2 * np.pi * f * n)

# Plot sinyal waktu kontinu dan diskrit
plt.figure(figsize=(10, 6))

# Plot sinyal kontinu
plt.subplot(2, 1, 1)
plt.plot(t, x_continuous)
plt.title('Sinyal Waktu Kontinu')
plt.xlabel('Waktu (detik)')
plt.ylabel('Amplitudo')

# Plot sinyal diskrit
plt.subplot(2, 1, 2)
plt.stem(n, x_discrete, 'r', basefmt=" ", use_line_collection=True)
plt.title('Sinyal Waktu Diskrit')
plt.xlabel('Waktu (detik)')
plt.ylabel('Amplitudo')

plt.tight_layout()
plt.show()
```
