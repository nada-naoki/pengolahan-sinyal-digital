```python
import numpy as np
import matplotlib.pyplot as plt

def H_ejw(omega):
    return 1 / (1 - 0.3 * np.exp(-1j * omega))

omega = np.linspace(0, np.pi, 500)
H_w = H_ejw(omega)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(omega / np.pi, np.abs(H_w), 'b')
plt.title('Magnitude Response of H(e^jw)')
plt.xlabel('Normalized Frequency (×π rad/sample)')
plt.ylabel('Magnitude')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(omega / np.pi, np.angle(H_w), 'b')
plt.title('Phase Response of H(e^jw)')
plt.xlabel('Normalized Frequency (×π rad/sample)')
plt.ylabel('Phase (radians)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
