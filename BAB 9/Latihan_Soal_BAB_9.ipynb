```python
import numpy as np
import matplotlib.pyplot as plt

fs = 1000 
duration = 3 
t = np.arange(0, duration, 1/fs)  
f_low = 5    
f_high = 20  
x_original = 4 * np.sin(2 * np.pi * f_low * t) + 8 * np.sin(2 * np.pi * f_high * t)

def downsample(x, M):
    return x[::M]
M = 5  
x_downsample = downsample(x_original,M)
t_downsample = t[:len(x_downsample)]  

plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
plt.plot(t, x_original)
plt.title('Sinyal Original')
plt.xlabel('Waktu [s]')
plt.ylabel('Amplitudo')

plt.subplot(2, 1, 2)
plt.plot(t_downsample, x_downsample)
plt.title('Sinyal Downsample')
plt.xlabel('Waktu [s]')
plt.ylabel('Amplitudo')

plt.tight_layout()
plt.show()
length_original = len(x_original)
length_downsample = len(x_downsample)
print (f"Panjang sinyal Asli = {length_original}")
print (f"Panjang sinyal Downsample = {length_downsample}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

L = 8
x_upsampled = []

for xi in x_downsample:
    x_upsampled.append(xi)
    x_upsampled.extend([0] * (L - 1))  

x_upsampled = np.array(x_upsampled)
t_upsampled = np.arange(0, len(x_upsampled)) * (t_downsample[1] - t_downsample[0])  

plt.figure(figsize=(8, 8))

plt.subplot(2, 1, 1)
plt.plot(t_downsample, x_downsample)
plt.title("Sinyal Downsampled")
plt.xlabel("Waktu [s]")
plt.ylabel("Amplitudo")
plt.subplot(2, 1, 2)
plt.plot(t_upsampled, x_upsampled)
plt.title("Sinyal Setelah Upsampling (L = 8)")
plt.xlabel("Waktu [s]")
plt.ylabel("Amplitudo")

plt.tight_layout()
plt.show()

print(f"Panjang sinyal Downsampled = {len(x_downsample)}")
print(f"Panjang sinyal Upsampled = {len(x_upsampled)}")
```
