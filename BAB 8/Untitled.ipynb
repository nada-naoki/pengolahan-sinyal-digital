```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Definisikan koefisien filter IIR (contoh butterworth lowpass filter)
b, a = signal.butter(4, 0.2)  # Orde 4 dan frekuensi cutoff 0.2*Nyquist

# Hitung respon impuls filter IIR
impulse = np.zeros(20)  # Sinyal impuls, panjang 20 sampel
impulse[0] = 1  # Impuls unit di n=0
h = signal.lfilter(b, a, impulse)  # Respon impuls filter IIR

# Plot respon impuls
plt.stem(np.arange(20), h, use_line_collection=True)
plt.title('Respon Impuls Filter IIR')
plt.xlabel('n')
plt.ylabel('h[n]')
plt.grid()
plt.show()
```
