```python
import numpy as np
import matplotlib.pyplot as plt

def y_n(n):
    return 5 * (0.5 ** n) - 5 * (0.3 ** n)
n = np.arange(0, 21)
y_vals = y_n(n)

plt.figure(figsize=(5, 5))
plt.stem(n, y_vals, basefmt=" ")
plt.title('$y[n] = 5 \cdot (0.5)^n - 5 \cdot (0.3)^n$')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.grid(True)
plt.show()
```
