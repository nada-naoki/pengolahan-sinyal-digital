```python
import numpy as np
import matplotlib.pyplot as plt

def solve_quadratic(a, b, c):
    discriminant = b**2 - 4*a*c
    root1 = (-b + np.sqrt(discriminant)) / (2*a)
    root2 = (-b - np.sqrt(discriminant)) / (2*a)
    return root1, root2

numerator = [1, -0.5]  # Pembilang (z - 0.5)
denominator = [1, -0.7, 0.12]  # Penyebut (z^2 - 0.7z + 0.12)

zeros = np.roots(numerator)
poles = solve_quadratic(denominator[0], denominator[1], denominator[2])

plt.figure(figsize=(5, 5))
plt.scatter(np.real(zeros), np.imag(zeros), s=100, marker='o', label='Zeros', color='blue')
plt.scatter(np.real(poles), np.imag(poles), s=100, marker='x', label='Poles', color='red')

unit_circle = plt.Circle((0, 0), 1, color='black', fill=False, linestyle='--')
plt.gca().add_artist(unit_circle)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.title('Zero-Pole Plot of H(z)')
plt.grid(True)
plt.legend()
plt.show()

print("Zeros:", zeros)
print("Poles:", poles)
if np.all(np.abs(poles) < 1):
    print("Sistem stabil (semua poles berada dalam unity circle).")
else:
    print("Sistem tidak stabil.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def H_ejw(omega):
    z = np.exp(1j * omega)
    numerator = 1 - 0.5*z**(-1)
    denominator = 1 - 0.7*z**(-1) + 0.12*z**(-2)
    return numerator / denominator

omega = np.linspace(0, np.pi, 500)
H_w = H_ejw(omega)


plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(omega / np.pi, np.abs(H_w), 'b')
plt.title('Magnitude Response of H(e^jw)')
plt.xlabel('Normalized Frequency ( rad/sample)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.subplot(1, 2, 2)
plt.plot(omega / np.pi, np.angle(H_w), 'b')
plt.title('Phase Response of H(e^jw)')
plt.xlabel('Normalized Frequency ( rad/sample)')
plt.ylabel('Phase (radians)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def H_ejw(omega):
    z = np.exp(1j * omega)
    numerator = 1 - 0.5 * z**(-1)
    denominator = 1 - 0.7 * z**(-1) + 0.12 * z**(-2)
    return numerator / denominator

# Define omega range from 0 to pi with 500 points
omega = np.linspace(0, np.pi, 500)
H_w = H_ejw(omega)

# Plot magnitude and phase responses
plt.figure(figsize=(10, 5))

# Magnitude response plot
plt.subplot(1, 2, 1)
plt.plot(omega / np.pi, np.abs(H_w), 'b')
plt.title('Magnitude Response of H(e^jw)')
plt.xlabel('Normalized Frequency ( rad/sample)')
plt.ylabel('Magnitude')
plt.grid(True)

# Phase response plot
plt.subplot(1, 2, 2)
plt.plot(omega / np.pi, np.angle(H_w), 'b')
plt.title('Phase Response of H(e^jw)')
plt.xlabel('Normalized Frequency ( rad/sample)')
plt.ylabel('Phase (radians)')
plt.grid(True)

plt.tight_layout()
plt.show()

# Define the specific points in normalized frequency (0, 0.25, 0.5, and 1)
freq_points = [0, 0.25, 0.5, 1]
for freq in freq_points:
    omega_point = freq * np.pi
    H_at_omega = H_ejw(omega_point)
    magnitude = np.abs(H_at_omega)
    phase = np.angle(H_at_omega)
    print(f"Frequency = {freq}: Magnitude = {magnitude:.4f}, Phase = {phase:.4f} radians")
```
