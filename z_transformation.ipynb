```python
import numpy as np
import matplotlib.pyplot as plt

def x_n(n):
    return 0.5**n * (n >= 0) 
def h_n(n):
    return 0.3**n * (n >= 0)

n = np.arange(0, 21)
x_vals = x_n(n)
h_vals = h_n(n)


plt.figure(figsize=(5, 5))
plt.stem(n, h_vals, basefmt=" ", linefmt='r-', markerfmt='ro')
plt.title('$h[n] = 0.3^n u(n)$')
plt.xlabel('n')
plt.ylabel('h[n]')
plt.grid(True)
plt.show()

zero = [0]  # Zero di z = 0
pole = [0.3]  # Pole di z = 0.3

plt.figure(figsize=(5,5))
plt.scatter(np.real(zero), np.imag(zero), s=100, marker='o', label='Zeros', color='blue')
plt.scatter(np.real(pole), np.imag(pole), s=100, marker='x', label='Poles', color='red')
unit_circle = plt.Circle((0, 0), 1, color='black', fill=False, linestyle='--')
plt.gca().add_artist(unit_circle)
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.title('Zero-Pole Plot of H(z)')
plt.grid(True)
plt.legend()
plt.show()
```
