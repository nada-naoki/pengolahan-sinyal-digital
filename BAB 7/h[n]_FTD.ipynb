```python
import numpy as np
import matplotlib.pyplot as plt

# Definisikan frekuensi cutoff dan jumlah sampel
Omega_C = np.pi / 4  # Frekuensi cutoff
M = 10  # Panjang filter (M)

# Definisikan n dari -M hingga M
n = np.arange(-M, M+1)

# Hitung respon impuls h[n]
h = np.zeros_like(n, dtype=float)
h[n == 0] = Omega_C / np.pi  # Ketika n = 0
h[n != 0] = np.sin(Omega_C * n[n != 0]) / (np.pi * n[n != 0])  # Ketika n != 0

# Plot respon impuls h[n]
plt.figure(figsize=(10, 6))
plt.stem(n, h, basefmt=" ", use_line_collection=True)
plt.title(r'Respon Impuls $h[n]$ untuk Low Pass Filter', fontsize=14)
plt.xlabel(r'$n$', fontsize=14)
plt.ylabel(r'$h[n]$', fontsize=14)
plt.grid(True)
plt.show()
```
