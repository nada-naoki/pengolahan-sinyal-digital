```python
import numpy as np
import matplotlib.pyplot as plt

# Definisikan frekuensi cutoff ter-normalisasi
fc = 800  # Frekuensi cutoff
fs = 8000  # Laju sampling
Omega_c = 2 * np.pi * fc / fs  # Cutoff ter-normalisasi

# Hitung koefisien filter
M = 1
h = np.zeros(2 * M + 1)
h[1] = Omega_c / np.pi  # h[0] untuk n = 0
h[2] = h[0] = np.sin(Omega_c) / np.pi  # h[1] untuk n = 1 dan h[-1] untuk n = -1

# Plot Respon Frekuensi Ideal (yang benar)
plt.figure(figsize=(10, 4))

# Respon Frekuensi Ideal
plt.subplot(1, 2, 1)
freq = np.linspace(-2 * np.pi, 2 * np.pi, 500)
ideal_response = np.where(np.abs(freq) <= Omega_c, 1, 0)  # Menggunakan Omega_c = 0.2pi
plt.plot(freq, ideal_response, 'b-', linewidth=2)
plt.title('Ideal Frequency Response')
plt.xlabel('Frequency (radians)')
plt.ylabel('Magnitude')
plt.grid(True)

# Plot Respon Impuls h[n]
plt.subplot(1, 2, 2)
plt.stem([-1, 0, 1], h, basefmt=" ", use_line_collection=True)
plt.title('Impulse Response h[n]')
plt.xlabel('n (Sample)')
plt.ylabel('Amplitude')
plt.grid(True)

# Tampilkan grafik
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definisikan frekuensi sudut
Omega = np.linspace(-3, 3, 500)  

# Inisialisasi array untuk magnitudo respon frekuensi
H_magnitude = np.zeros(len(Omega))

# Menghitung magnitudo respon frekuensi 
for i in range(len(Omega)):
    H_magnitude[i] = 0.2 + 0.3742 * np.cos(Omega[i])

# Plot Magnitude Frequency Response
plt.figure(figsize=(6, 4))
plt.plot(Omega, H_magnitude, 'b-', linewidth=2)  # Skala frekuensi diubah ke -3 hingga 3
plt.title('Frequency Response (Truncated)')
plt.xlabel('Ohm') 
plt.ylabel('Magnitude')
plt.grid(True)

# Tampilkan grafik
plt.show()
```
