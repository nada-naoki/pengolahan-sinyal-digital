```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the frequency cutoff and the frequency domain
Omega_C = np.pi / 4  # Cutoff frequency
Omega = np.linspace(-2 * np.pi, 2 * np.pi, 2000)

# Define the ideal frequency response for LPF in repeated intervals
H = np.zeros_like(Omega)
for i in range(-2, 3):
    condition = np.logical_and(Omega >= i * 2 * np.pi - Omega_C, Omega <= i * 2 * np.pi + Omega_C)
    H[condition] = 1

# Plot the repeated ideal LPF response
plt.figure(figsize=(10, 6))
plt.plot(Omega, H, label=r'$H(e^{j\Omega})$', linewidth=2)
plt.title('Repeated Ideal Low Pass Filter Response')
plt.xlabel(r'$\Omega$ (Frequency)', fontsize=14)
plt.ylabel(r'$H(e^{j\Omega})$', fontsize=14)
plt.ylim([-0.1, 1.2])
plt.xlim([-2 * np.pi, 2 * np.pi])
plt.grid(True)
plt.show()

# Now for the single period graph (-pi to pi)
Omega_single = np.linspace(-np.pi, np.pi, 1000)

# Define the ideal frequency response for LPF for a single period
H_single = np.where(np.abs(Omega_single) <= Omega_C, 1, 0)

# Plot the single period ideal LPF response
plt.figure(figsize=(8, 6))
plt.plot(Omega_single, H_single, label=r'$H(e^{j\Omega})$', linewidth=2)
plt.title('Ideal Low Pass Filter Response (Single Period)')
plt.xlabel(r'$\Omega$ (Frequency)', fontsize=14)
plt.ylabel(r'$H(e^{j\Omega})$', fontsize=14)
plt.ylim([-0.1, 1.2])
plt.xlim([-np.pi, np.pi])
plt.grid(True)
plt.legend()
plt.show()
```
