```python
import numpy as np
import matplotlib.pyplot as plt

# Membuat sinyal dengan noise
np.random.seed(42)  # Untuk memastikan hasil yang sama
n = 500
original_signal = np.concatenate((np.zeros(150), np.ones(100), np.zeros(250)))  # Sinyal asli
noise = np.random.normal(0, 0.3, n)  # Noise acak
noisy_signal = original_signal + noise  # Sinyal dengan noise

# Fungsi untuk menghitung Moving Average
def moving_average(signal, M):
    N = len(signal)
    output = np.zeros(N - M + 1)
    for i in range(len(output)):
        output[i] = np.sum(signal[i:i+M]) / M
    return output

# Menghitung Moving Average dengan berbagai ukuran jendela
M_values = [11, 51]
filtered_signals = [moving_average(noisy_signal, M) for M in M_values]

# Plot sinyal asli, sinyal dengan noise, dan hasil filter
plt.figure(figsize=(10, 8))

# Plot sinyal asli dengan noise (Panel a)
plt.subplot(3, 1, 1)
plt.plot(noisy_signal, label='Sinyal dengan Noise', color='black')
plt.title('a. Sinyal dengan Noise')
plt.ylabel('Amplitudo')
plt.grid(True)

# Plot sinyal setelah Moving Average Filter 11 point (Panel b)
plt.subplot(3, 1, 2)
plt.plot(np.arange(len(filtered_signals[0])) + (M_values[0]-1)//2, filtered_signals[0], label=f'M = {M_values[0]}', color='black')
plt.title(f'b. 11 point Moving Average')
plt.ylabel('Amplitudo')
plt.grid(True)

# Plot sinyal setelah Moving Average Filter 51 point (Panel c)
plt.subplot(3, 1, 3)
plt.plot(np.arange(len(filtered_signals[1])) + (M_values[1]-1)//2, filtered_signals[1], label=f'M = {M_values[1]}', color='black')
plt.title(f'c. 51 point Moving Average')
plt.ylabel('Amplitudo')
plt.xlabel('Nomor Sampel')
plt.grid(True)

plt.tight_layout()
plt.show()
```
